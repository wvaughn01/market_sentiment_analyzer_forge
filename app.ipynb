```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib

model = joblib.load("behavioral_model.pkl")

def predict_reaction(model, new_data):
    prediction = model.predict(new_data)[0]
    reaction_map = {
        0: "ðŸ“‰ Negative Reaction â€“ Expect market downturn",
        1: "ðŸ“Š Neutral Reaction â€“ Little or no movement",
        2: "ðŸ“ˆ Positive Reaction â€“ Expect market uptrend"
    }
    return prediction, reaction_map.get(prediction, "Unknown")

st.set_page_config(page_title="Market Reaction Predictor", layout="centered")
st.title("ðŸ§  Behavioral Market Reaction Predictor")
st.markdown("Enter mock or real market features to predict expected market behavior.")

col1, col2 = st.columns(2)
with col1:
    price_change = st.number_input("Price Change (%)", value=0.5)
    volatility = st.number_input("Volatility", value=1.2)
    sentiment_score = st.number_input("Sentiment Score", value=0.3)
with col2:
    loss_aversion_score = st.number_input("Loss Aversion Score", value=0.6)
    herding_index = st.number_input("Herding Index", value=0.4)
    reaction_speed = st.number_input("Reaction Speed", value=0.7)

if st.button("ðŸ”® Predict Market Reaction"):
    input_df = pd.DataFrame([{
        'price_change': price_change,
        'volatility': volatility,
        'sentiment_score': sentiment_score,
        'loss_aversion_score': loss_aversion_score,
        'herding_index': herding_index,
        'reaction_speed': reaction_speed
    }])
    prediction, explanation = predict_reaction(model, input_df)
    st.success(f"**Prediction:** {explanation}")
```
